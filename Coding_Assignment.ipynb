```python
import pandas as pd
```

```python
data = {
"Roll No":[1,1,2,2,1,2] , 
"Subject":['DS','Java','DS','Java','DS','Java'], 
"Marks":[25,40,30,26,40,50], 
"Date": ["01-01-23","02-01-23","01-01-23","02-01-23","15-01-23","15-01-23"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Date']= pd.to_datetime(df['Date'], format="%d-%m-%y")
```

```python
df.info()
```

```python
df_sorted = df.sort_values(
    by=['Roll No', 'Subject', 'Date'], 
    ascending=[True, True, False]
)
```

```python
def get_latest_marks(group):
    marks = group['Marks'].head(3).tolist()
    
    marks.extend([0] * (3 - len(marks)))
    
    latest_date = group['Date'].iloc[0]
    
    return pd.Series({
        'M1': marks[0],
        'M2': marks[1],
        'M3': marks[2],
        'Date': latest_date
    })
```

```python
df_output = df_sorted.groupby(['Roll No', 'Subject']).apply(get_latest_marks).reset_index()
```

```python
df_output = df_output.rename_axis(None, axis=1)
df_final = df_output[['Roll No', 'Subject', 'M1', 'M2', 'M3', 'Date']]
```

```python
df_final.to_csv('output_file.csv', index=False)
```
